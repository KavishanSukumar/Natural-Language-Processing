```python
import nltk
from nltk.stem import PorterStemmer
from nltk import FreqDist
import re
from functools import reduce
from nltk.stem import WordNetLemmatizer
 
lemmatizer = WordNetLemmatizer()
ps = PorterStemmer()
```

```python
file12=nltk.corpus.gutenberg.fileids()[12]
mobydicktext=nltk.corpus.gutenberg.raw(file12)
mobydicktokens=nltk.wordpunct_tokenize(mobydicktext)
mobydicktokens
```

```python
stemmedtokens = [ps.stem(token) for token in mobydicktokens]
```

```python
print(stemmedtokens[:50])
```

```python
lemmtizedtokens=[lemmatizer.lemmatize(token) for token in mobydicktokens]
```

```python
print(lemmtizedtokens[:50])
```
